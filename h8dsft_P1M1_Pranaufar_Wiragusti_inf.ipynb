```python
# Import Library

import pickle
import json
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report, precision_score, recall_score, accuracy_score, f1_score, ConfusionMatrixDisplay
```

```python

# Save Random Forest model
rfs = 'random_forest.pkl'
rf = pickle.load(open(rfs, 'rb'))

scalers = 'scaler.pkl'
scaler = pickle.load(open(scalers,'rb'))

list_cols = 'list_cols.pkl'
json.load(open(list_cols,'r'))
```

```python
df = pd.read_csv('inferrence.csv')
df
```

```python
default = df['default_payment_next_month']
df.drop(['Unnamed: 0','default_payment_next_month'],axis=1,inplace=True)
```

```python
# Predict 
dfscaled = scaler.transform(df)
y_pred_inf = rf.predict(dfscaled)
y_pred_inf
```

```python
test = pd.DataFrame()
test['truth'] = default
test['predicition'] = y_pred_inf

test.replace({0:'Not Default',1:'Default'},inplace=True)
test
```

```python
test['truth'].value_counts()
```

```python
test['predicition'].value_counts()
```

```python
print('Accuracy:', accuracy_score(default, y_pred_inf))
print('Confusion Matrix:\n', confusion_matrix(default, y_pred_inf))
print('Classification Report:\n', classification_report(default, y_pred_inf))
```
